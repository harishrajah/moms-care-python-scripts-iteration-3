```python
# Importing necessary packages

import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
#!pip install pdftables.six
from pdftables import get_tables
```

```python
# opening the pdf file and using get_tables function
# The get_tables() function returns the page as its own table
# The tables have a list of rows, and each of those row is a contained list of columns

pdf_file = 'language_chinese_suburb.pdf'
pdf_obj = open(pdf_file, 'rb')
tables = get_tables(pdf_obj)
```

```python
# print out each row in table
# creating a DataFrame by passing a dictionary of objects

df = {}
for table in tables:
    for row in table[8:28]:
        df[row[0]] = row[1:]
        print (row)

df = pd.DataFrame(df)
df
```

```python
# The last step is to transpose the DataFrame so that each row is a record for a council
# save the data into a CSV file

df = df.T
df.columns = range(14)
df.to_csv('language.csv')
```
